```python
# Bài 1
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 10)
y = x**2 + 2*x
plt.plot(x, y, label='y = x^2 + 2x')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Biểu đồ hàm số y=x^2+2x')
plt.legend()
plt.grid()
plt.show()
```

```python
# Bài 4
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 200)
y1 = x
y2 = x**2
y3 = x**3

plt.plot(x, y1, label="y = x")
plt.plot(x, y2, label="y = x^2")
plt.plot(x, y3, label="y = x^3")

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("So sánh 3 hàm y = x, y = x^2, y = x^3")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Bài 5
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

x = np.linspace(-5, 5, 200)
axes[0, 0].plot(x, x)
axes[0, 0].set_title("y = x")

axes[1, 0].plot(x, x**2)
axes[1, 0].set_title("y = x^2")

x_rand = np.random.randint(0, 50, 100)
y_rand = np.random.randint(0, 50, 100)
axes[0, 1].scatter(x_rand, y_rand)
axes[0, 1].set_title("Scatter ngẫu nhiên phân tán")

x_cluster = np.random.normal(25, 5, 100)
y_cluster = np.random.normal(25, 5, 100)
axes[1, 1].scatter(x_cluster, y_cluster)
axes[1, 1].set_title("Scatter phân bố cụm")

plt.tight_layout()
plt.show()
```

```python
# Bài 6
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 300)
y = np.sin(x)

plt.plot(x, y)
point_x = np.pi / 2
point_y = np.sin(point_x)
plt.scatter(point_x, point_y, color='red')
plt.annotate("Cực đại: (π/2, 1)", xy=(point_x, point_y),
             xytext=(point_x+0.5, point_y-0.5),
             arrowprops=dict(arrowstyle="->"))

plt.xlabel("x")
plt.ylabel("sin(x)")
plt.title("Đồ thị y = sin(x) và chú thích cực đại")
plt.grid(True)
plt.show()
```

```python
# Bài 8
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

x = np.linspace(0, 2*np.pi, 200)
y = np.sin(x)

fig, ax = plt.subplots()
line, = ax.plot(x, y)
ax.set_ylim(-1.5, 1.5)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("Animation: y = sin(x) thay đổi theo thời gian")

def update(frame):
    line.set_ydata(np.sin(x + frame/10))
    return line,

ani = FuncAnimation(fig, update, frames=100, interval=50, blit=True)
plt.show()
```

```python
# Bài 9
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(0)
days = pd.date_range("2025-12-01", periods=30)
temperature = 20 + np.random.randn(30).cumsum() * 0.5

fig, ax = plt.subplots(figsize=(12,6))
ax.plot(days, temperature, label="Nhiệt độ hàng ngày")

z = np.polyfit(range(len(temperature)), temperature, 1)
trend = np.polyval(z, range(len(temperature)))
ax.plot(days, trend, color='orange', linestyle='--', label="Trendline")

spike_threshold = 1.5 
spikes = np.abs(temperature - trend) > spike_threshold
ax.scatter(days[spikes], temperature[spikes], color='red', label="Spikes")

special_dates = [days[14], days[24]]
ax.scatter(special_dates, temperature[[14, 24]], color='green', label="Ngày đặc biệt", s=100)

ax.set_xlabel("Ngày")
ax.set_ylabel("Nhiệt độ (°C)")
ax.set_title("Phân tích dữ liệu nhiệt độ hàng ngày trong tháng")
ax.legend()
ax.grid(True)
plt.xticks(rotation=45)
plt.show()
```
