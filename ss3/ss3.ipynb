```python
# Bài 1
number_1 = int(input("Nhập số thứ nhất: "))
number_2 = int(input("Nhập số thứ hai: "))

def difference(a, b):
    return a - b

def multiply(a, b):
    return a * b

print("Hiệu của hai số là:", difference(number_1, number_2))
print("Tích của hai số là:", multiply(number_1, number_2))
```

```python
# Bài 2
number = int(input("Nhập một số nguyên: "))

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

print(f"Số {number} là số nguyên tố:" if is_prime(number) else f"Số {number} không phải là số nguyên tố.")
```

```python
# Bài 3
number = int(input("Nhập một số nguyên: "))

def factorial(n):
    if n < 0:
        return "Không có giai thừa cho số âm"
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result

print(f"Giai thừa của {number} là:", factorial(number))
```

```python
# Bài 4
import math
number = int(input("Nhập một số nguyên dương: "))

def circle_area(r):
    return math.pi * r * r

print(f"Diện tích hình tròn với bán kính {number} là:", round(circle_area(number), 2))
```

```python
# Bài 5
def is_leap_year(year):
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True
    return False

year = int(input("Nhập một năm: "))
print(f"Năm {year} là năm nhuận." if is_leap_year(year) else f"Năm {year} không phải là năm nhuận.")
```

```python
# Bài 6
number = int(input("Nhập một số nguyên dương: "))

def sum_of_integers(n):
    total = 0
    for digit in str(n):
        total += int(digit)
    return total

print(f"Tổng các chữ số của {number} là:", sum_of_integers(number))
```

```python
# Bài 7
number = int(input("Nhập một số nguyên dương: "))

def reverse_integer(n):
    return int(str(n)[::-1])

print("Số nguyên đảo ngược là:", reverse_integer(number))
```

```python
# Bài 8

number = int(input("Nhập một số nguyên dương: "))

def is_perfect(n):
    if n < 2:
        return False
    divisors_sum = sum(i for i in range(1, n) if n % i == 0)
    return divisors_sum == n

print(f"Số {number} là số hoàn hảo." if is_perfect(number) else f"Số {number} không phải là số hoàn hảo.")
```
