```python
import re

# Bài tập tổng hợp - products là list của tuples (id, name, price)

products = []

def is_int(s):
    return bool(re.match(r'^[+-]?\d+$', s.strip()))

def is_float(s):
    return bool(re.match(r'^[+-]?(\d+(\.\d*)?|\.\d+)$', s.strip()))

def add_product(name, price):
    if any(p[1] == name for p in products):
        print(f"Sản phẩm {name} đã tồn tại.")
        return
    new_id = max((p[0] for p in products), default=0) + 1
    products.append((new_id, name, price))
    print(f"Đã thêm sản phẩm {name} với giá {price} (ID: {new_id}).")

def show_products():
    if not products:
        print("Chưa có sản phẩm nào.")
        return
    for pid, name, price in products:
        print(f"ID: {pid}, Tên: {name}, Giá: {price}")

def find_product_by_name(name):
    for pid, n, price in products:
        if n == name:
            print(f"Tìm thấy sản phẩm: ID: {pid}, Tên: {n}, Giá: {price}")
            return
    print(f"Không tìm thấy sản phẩm với tên {name}.")

def update_product(product_id, new_name=None, new_price=None):
    for i, (pid, n, price) in enumerate(products):
        if pid == product_id:
            updated_name = new_name if new_name is not None else n
            updated_price = new_price if new_price is not None else price
            products[i] = (pid, updated_name, updated_price)
            print(f"Đã cập nhật sản phẩm ID {product_id} thành Tên: {updated_name}, Giá: {updated_price}.")
            return
    print(f"Không tìm thấy sản phẩm với ID {product_id}.")

def delete_product(product_id):
    global products
    if any(p[0] == product_id for p in products):
        products = [p for p in products if p[0] != product_id]
        print(f"Đã xoá sản phẩm với ID {product_id}.")
    else:
        print(f"Không tìm thấy sản phẩm với ID {product_id}.")

while True:
    print("Menu")
    print("1. Thêm sản phẩm")
    print("2. Hiển thị sản phẩm")
    print("3. Tìm sản phẩm theo tên")
    print("4. Cập nhật sản phẩm")
    print("5. Xoá sản phẩm")
    print("0. Thoát")
    choice = input("Chọn một tùy chọn (1-5), 0 để thoát: ").strip()
    if not is_int(choice):
        print("Lựa chọn không hợp lệ. Vui lòng chọn lại.")
        continue
    choice = int(choice)
    if choice < 0 or choice > 5:
        print("Lựa chọn không hợp lệ. Vui lòng chọn lại.")
        continue
    if choice == 1:
        name = input("Nhập tên sản phẩm: ")
        price_input = input("Nhập giá sản phẩm: ").strip()
        if not is_float(price_input):
            print("Giá không hợp lệ.")
            continue
        price = float(price_input)
        add_product(name, price)
    elif choice == 2:
        show_products()
    elif choice == 3:
        name = input("Nhập tên sản phẩm cần tìm: ")
        find_product_by_name(name)
    elif choice == 4:
        product_id_input = input("Nhập ID sản phẩm cần cập nhật: ").strip()
        if not is_int(product_id_input):
            print("ID không hợp lệ.")
            continue
        product_id = int(product_id_input)
        new_name = input("Nhập tên mới (để trống nếu không đổi): ")
        new_price_input = input("Nhập giá mới (để trống nếu không đổi): ").strip()
        if new_price_input and not is_float(new_price_input):
            print("Giá không hợp lệ.")
            continue
        new_price = float(new_price_input) if new_price_input else None
        update_product(product_id, new_name if new_name else None, new_price)
    elif choice == 5:
        product_id_input = input("Nhập ID sản phẩm cần xoá: ").strip()
        if not is_int(product_id_input):
            print("ID không hợp lệ.")
            continue
        product_id = int(product_id_input)
        delete_product(product_id)
    elif choice == 0:
        print("Thoát chương trình.")
        break
    else:
        print("Lựa chọn không hợp lệ.")
```
