```python
# Bài 8
products = {"A": 120, "B": 80, "C": 300, "D": 50}

print(f"Sản phẩm A có giá {products.get('A')}" if products.get('A') is not None else "A đã bị xoá")
print(f"Mảng sau khi được sắp xếp tăng dần theo giá: {dict(sorted(products.items(), key=lambda item: item[1]))}")
print(f"Tổng giá trị của tất cả sản phẩm: {sum(products.values())}")

deleted_higher_than_200 = {k: v for k, v in products.items() if v <= 200}

print(f"Danh sách sản phẩm sau khi xoá các sản phẩm có giá cao hơn 200: {deleted_higher_than_200}")
```

```python
# Bài tập tổng hợp
products = []

def add_product(name, price):
    if any(p['name'] == name for p in products):
        print(f"Sản phẩm {name} đã tồn tại.")
        return
    products.append({'id': len(products) + 1, 'name': name, 'price': price})
    print(f"Đã thêm sản phẩm {name} với giá {price}.")

def show_products():
    if not products:
        print("Chưa có sản phẩm nào.")
        return
    for p in products:
        print(f"ID: {p['id']}, Tên: {p['name']}, Giá: {p['price']}")

def find_product_by_name(name):
    for p in products:
        if p['name'] == name:
            print(f"Tìm thấy sản phẩm: ID: {p['id']}, Tên: {p['name']}, Giá: {p['price']}")
            return
    print(f"Không tìm thấy sản phẩm với tên {name}.")

def update_product(product_id, new_name = None, new_price = None):
    for p in products:
        if p['id'] == product_id:
            p['name'] = new_name
            p['price'] = new_price
            print(f"Đã cập nhật sản phẩm ID {product_id} thành Tên: {new_name}, Giá: {new_price}.")
            return
    print(f"Không tìm thấy sản phẩm với ID {product_id}.")

def delete_product(product_id):
    global products
    products = [p for p in products if p['id'] != product_id]
    print(f"Đã xoá sản phẩm với ID {product_id}.")

while True:
    print("Menu")
    print("1. Thêm sản phẩm")
    print("2. Hiển thị sản phẩm")
    print("3. Tìm sản phẩm theo tên")
    print("4. Cập nhật sản phẩm")
    print("5. Xoá sản phẩm")
    print("0. Thoát")
    choice = int(input("Chọn một tùy chọn (1-6), 0 để thoát: "))
    if choice == 1:
        name = input("Nhập tên sản phẩm: ")
        price = float(input("Nhập giá sản phẩm: "))
        add_product(name, price)
    elif choice == 2:
        show_products()
    elif choice == 3:
        name = input("Nhập tên sản phẩm cần tìm: ")
        find_product_by_name(name)
    elif choice == 4:
        product_id = int(input("Nhập ID sản phẩm cần cập nhật: "))
        new_name = input("Nhập tên mới (để trống nếu không đổi): ")
        new_price_input = input("Nhập giá mới (để trống nếu không đổi): ")
        new_price = float(new_price_input) if new_price_input else None
        update_product(product_id, new_name if new_name else None, new_price)
    elif choice == 5:
        product_id = int(input("Nhập ID sản phẩm cần xoá: "))
        delete_product(product_id)
    elif choice == 0:
        print("Thoát chương trình.")
        break
    
```
